```python
from tkinter import *
import math
textLB = ''
so1 = 0
so2 = 0
dau = ''
def charr(event):
    global so1
    global so2
    global dau
    global textLB
    text = event.widget['text']
    if text in so:
        if (textLB !='' and text =='0') or text !='0':
            textLB+=text
        else:
            textLB=''
        display()
    if text in dauu:
        dau = text
        so1 = float(label.get())
        textLB=''
        display()      
    if text == 'C':
        textLB = ''
        so1 = 0
        so2 = 0
        dau = ''
        display()
    if text == 'CE':
        if dau == '':
            so1 = 0
        else:
            so2 = 0
        textLB = ''
        display()
    if text == "<--":
        if textLB=='':
            return
        else:
            textLB = textLB[0:len(textLB)-1]
            display()
    if text == "1/x":
        textLB = label.get()
        if textLB == '':
            textLB = 'Cannot divide by zero';
        else:
            textLB = 1/float(textLB)
        label.delete(0,'end')
        label.insert(0,textLB)
        textLB=''
    if text == '%':
        textLB = label.get()
        if textLB == '':
            textLB = '0'
        else:
            textLB = float(textLB)/100
        label.delete(0,'end')
        label.insert(0,textLB)
        textLB=''
    if text == 'sqrt':
        textLB = label.get()
        if textLB == '':
            return
        else:
            textLB = math.sqrt(float(textLB))
        label.delete(0,'end')
        label.insert(0,textLB)
        textLB=''
        
    if text == '+-':
        textLB = label.get()
        if textLB != '':
            num = float(textLB)
            num =-num
            textLB=str(num)
            label.delete(0,'end')
            label.insert(0,textLB)
            textLB=''
        
        
    global memory
    if text in l3:
        if text=='MS':
            memory=float(label.get())
        elif text=='MR':
            textLB=str(memory)
            display()
            textLB=''
        elif text=='M+':
            memory+=float(label.get())
        elif text=='M-':
            memory-=float(label.get())
        else:
            memory=0
def display():
    label.delete(0,'end')
    if textLB == '':
        label.insert(0,'0')
    else:
        label.insert(0,textLB)
    
def bang(event):
    global so2
    global textLB
    text = event.widget['text']
    so2 = float(label.get())
    if dau == '+':
        textLB=so1+so2
    if dau == '-':
        textLB=so1-so2
    if dau == '*':
        textLB=so1*so2
    if dau == '/':
        if so2 == 0:
            textLB ="NaN"
        else:
            textLB=so1/so2
    label.delete(0,'end')
    label.insert(0,textLB)
    textLB=''
    
    
window = Tk()
window.title("Calculator")

label = Entry(window,text="0",borderwidth=2,relief="solid",width=40, justify="right")
label.grid(row=0,column=1,columnspan = 5, padx = 5, pady = 5)

list = ["MC","MR","MS","M+","M-","<--","CE","C","+-","sqrt","7","8","9","/","%","4","5","6","*","1/x"]
l = ["1","2","3","-"]
l2 = [".","+"]
l3 = ["MC","MR","MS","M+","M-"]
so = '1234567890.'
dauu = '+*/-'
d = 0
r = 1
c = 1
for i in range(0,len(list)):
    bt = Button(window, text=list[i],width=5)
    bt.grid(row=r,column=c, padx = 5, pady = 5)
    d+=1
    c+=1
    if d%5==0:
        r+=1
        c =1
    bt.bind("<Button-1>",charr)
    
for i in range(0,len(l)):
    bt = Button(window, text=l[i],width=5)
    bt.grid(row=5,column=i+1, padx = 5, pady = 5)
    bt.bind("<Button-1>",charr)
    
btBang = Button(window, text="=",width=5,height=3)
btBang.grid(row=5,column=5,rowspan=2, padx = 5, pady = 5)
btBang.bind("<Button-1>",bang)

bt0 = Button(window, text="0",width=13)
bt0.grid(row=6,column=1,columnspan=2, padx = 5, pady = 5)
bt0.bind("<Button-1>",charr) 

for i in range(0,len(l2)):
    bt = Button(window, text=l2[i],width=5)
    bt.grid(row=6,column=i+3, padx = 5, pady = 5)
    bt.bind("<Button-1>",charr)
# mytext = bt['text']
# print(mytext)
window.mainloop()
```
